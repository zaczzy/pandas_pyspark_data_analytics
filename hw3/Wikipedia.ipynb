```python
topics = wikipedia.search("Pennsylvania", 10)
topics.extend(wikipedia.search("Trump"))
topics.extend(wikipedia.search("Apple"))
topics.extend(wikipedia.search("Google"))
topics.extend(wikipedia.search("Farm"))
topics.extend(wikipedia.search("Climate"))
topics.extend(wikipedia.search("Cloud"))
topics.extend(wikipedia.search("Football"))
topics.extend(wikipedia.search("Government"))
topics.extend(wikipedia.search("Putin"))

topics
```

```python
docs = {}

for topic in topics:
    try:
        result = wikipedia.page(topic)
        print ('Topic', topic)
        file = open("text/" + topic + ".txt", 'w')
        file.write(result.content)
        docs[topic] = result.content
    except wikipedia.DisambiguationError:
        result = ''
```
