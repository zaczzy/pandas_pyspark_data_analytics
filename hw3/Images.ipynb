```python
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
real_pandas = ndimage.imread('panda-mania-12.jpg')
plt.imshow(real_pandas)
```

```python
def convert_to_greyscale(image, crop_left, crop_top, crop_right, crop_bottom, contrast_scale):
    crop_img = image[crop_top:-crop_bottom, crop_left:-crop_right]
    gray_img = np.average(crop_img, axis=2)
    sx, sy = gray_img.shape
    median = np.median(gray_img)
    for x in range(sx):
        for y in range(sy):
            value = gray_img[x,y]
            if value != median:
                gray_img[x,y] = value * (1 + contrast_scale) if value > median else value * (1 - contrast_scale)
                if gray_img[x,y] < 0:
                    gray_img[x,y] = 0
                elif gray_img[x,y] >255:
                    gray_img[x,y] = 255
    return gray_img
```

```python
data = ndimage.imread('data-behind-everything.jpg')
gray_data = convert_to_greyscale(data, 10, 50, 10, 50, 0.3)
```
