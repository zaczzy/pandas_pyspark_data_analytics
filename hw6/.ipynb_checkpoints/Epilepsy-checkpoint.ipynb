```python
import numpy as np

def fft(time_data):
    return np.log10(np.absolute(np.fft.rfft(time_data, axis=1)[:,1:48]))
```

## Step 1.3

```python
# For each segment in data.keys(), take data[segment], transpose it, and call fft() on it.
# Take the result (all channels by all 47 frequencies) and convert it into a 1D array.
# This is a feature row in 2D array X.
# Set the class label for the same row in 1D array y, dependong on if the segment name 
# includes 'interictal' (this is a non-match) vs 'ictal' (this is a match)
X = []
y = []
def class_label(segment_name):
    if 'interictal' in segment_name:
        return 0
    elif 'ictal' in segment_name:
        return 1
    else:
        return -1
for segment in data.keys():
    X.append(np.reshape(fft(data[segment].transpose()), -1).tolist())
    y.append(class_label(segment))
X = np.array(X)
y = np.array(y)
```

```python
# Create X_train, X_test, y_train, y_test with
# random_state = 42 and test_size = 0.3
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
mms = StandardScaler()
X_train_norm = mms.fit_transform(X_train)
X_test_norm = mms.fit_transform(X_test)
```

```python
# X_test_norm
```

## Step 1.3.1 Classifier Evaluation


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn.cross_validation import cross_val_score
# Results, as a list of dictionaries
classifier_results = []
```

```python
## depth-5 decision tree
dt_model = DecisionTreeClassifier(max_depth=5, random_state=42)
test_scores = cross_val_score(estimator=dt_model, X=X_train_norm, y=y_train,cv=5,n_jobs=2)
classifier_results.append({'Classifier': 'DecTree-Depth-5','CV accuracy scores': test_scores, 'CV accuracy': '%.3f +/- %.3f' % (np.mean(test_scores), np.std(test_scores))})
dt_model.fit(X_train_norm, y_train)
print(dt_model.score(X_test_norm, y_test))
```

```python
## log-reg-l1
lr_model = LogisticRegression(solver='liblinear', penalty='l1', random_state=42)
test_scores = cross_val_score(estimator=lr_model, X=X_train_norm, y=y_train,cv=5,n_jobs=2)
classifier_results.append({'Classifier': 'LogReg-L1', 'CV accuracy scores': test_scores, 'CV accuracy': '%.3f +/- %.3f' % (np.mean(test_scores), np.std(test_scores))})
lr_model.fit(X_train_norm, y_train)
print(lr_model.score(X_test_norm, y_test))
```

```python
## log-reg-l2
lr_model = LogisticRegression(solver='liblinear', penalty='l2', random_state=42)
test_scores = cross_val_score(estimator=lr_model, X=X_train_norm, y=y_train,cv=5,n_jobs=2)
classifier_results.append({'Classifier': 'LogReg-L2', 'CV accuracy scores': test_scores, 'CV accuracy': '%.3f +/- %.3f' % (np.mean(test_scores), np.std(test_scores))})
lr_model.fit(X_train_norm, y_train)
print(lr_model.score(X_test_norm, y_test))
```

```python
## SVM
svm_model = SVC(random_state=42)
test_scores = cross_val_score(estimator=svm_model, X=X_train_norm, y=y_train,cv=5,n_jobs=2)
classifier_results.append({'Classifier': 'SVM',  'CV accuracy scores': test_scores, 'CV accuracy': '%.3f +/- %.3f' % (np.mean(test_scores), np.std(test_scores))})
svm_model.fit(X_train_norm, y_train)
print(svm_model.score(X_test_norm, y_test))
```
